```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))

# pais 1
x = [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018] #aos
y = [77, 75, 70, 76, 69, 70, 75, 81, 91, 90, 79, 88, 82, 84, 78] #puntos

# pais 2
x2 = [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018] #aos
y2 = [45, 42, 44, 44, 45, 46, 44, 42, 44, 42, 43, 41, 43, 43, 44] #puntos

# plotting
plt.plot(x, y, marker='o', linestyle='--', color='g', label='Mayor puntuacin')
plt.plot(x2, y2, marker='d', linestyle='-', color='r', label='Menor puntuacin')

# agregar nombre a los ejes y titulo al grafico
plt.xlabel('Aos')

plt.title('Mayor puntuacion vs Menor puntuacin')
plt.legend(loc='lower right')
plt.ylabel('Puntos')


# guardar figura
plt.savefig('ejemplo.png')

# mostrar plot
plt.show()
```
